```python
import cv2
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image=cv2.imread('testting.png')
plt.figure(figsize=(10,10))
plt.imshow(image)
```

```python
plt.title('R channel')
plt.ylabel('Height {}'.format(image.shape[0]))
plt.xlabel('Width {}'.format(image.shape[1]))

plt.imshow(image[ : , : , 0])
plt.show()
```

```python
pic = cv2.imread('testting.png')

pic[20:80 , : , 0] = 255 
plt.figure( figsize = (10,10))
plt.imshow(pic)
plt.show()
```

```python
pic[100:200 , : , 1] = 255 
plt.figure( figsize = (10,10))
plt.imshow(pic)
plt.show()
```

```python
pic[250:300 , : , 2] = 255 # full intensity to those pixel's B channel
plt.figure( figsize = (10,10))
plt.imshow(pic)
plt.show()
```

```python
pic[ 50:100 , 150:200 , [0,1,2] ] = 200
plt.figure( figsize = (10,10))
plt.imshow(pic)
plt.show()
```

```python
pic = cv2.imread('testting.png')

fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), ax):
    
    # create zero matrix
    split_img = np.zeros(pic.shape, dtype="uint8") # 'dtype' by default: 'numpy.float64'
    
    # assing each channel 
    split_img[ :, :, c] = pic[ :, :, c]
    
    # display each channel
    ax.imshow(split_img)
```

```python
pic = cv2.imread('testting.png')

gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(pic)  

plt.figure( figsize = (10,10))
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```
