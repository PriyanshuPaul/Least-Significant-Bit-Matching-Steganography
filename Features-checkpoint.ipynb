```python
from PIL import Image

import numpy as np
```

```python
width, height = Image.open("lena_gray.bmp").size
```

```python
im = Image.open("lena_gray.bmp")
pix_val = list(im.getdata())
```

```python
pix_val
```

```python
l=len(pix_val)
a=np.array(pix_val)
```

```python
B = a.reshape(width,height)
```

```python
B
```

```python
def delta(s,t):
    if s == t:
        return 1
    else:
        return 0
```

```python

val = []
def h0(d):
    sum = 0
    for i in range(width-1):
        for j in range(height-1):
            fij = B[i,j]
            sum += delta(fij,d)
    v = sum / (width * height)
    val.append(v)
```

```python
for d in range(256):
    h0(d)
```
